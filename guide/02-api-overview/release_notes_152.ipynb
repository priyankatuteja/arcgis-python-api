# What's new in version 1.5.2

#### New

#### Fixes
* Fixed `map widget` so default basemap draws if `gis` has not been set
* Fixed [`import_data()`]() function to publish Spatially Enabled DataFrames as Feature Layers
* Added new members to [`UserManager`](https://developers.arcgis.com/python/api-reference/arcgis.gis.toc/#usermanager) objects for support of the new [`User Type licensing`](https://www.esri.com/arcgis-blog/products/arcgis-online/announcements/whats-coming-in-arcgis-online-introducing-user-types/) model
* Added a `get_unique_values()` function to [`FeatureLayer`](https://developers.arcgis.com/python/api-reference/arcgis.features.toc/#featurelayer) objects
* Added batch sharing capability to the [`ContentManager`](https://developers.arcgis.com/python/api-reference/arcgis.gis.toc/#contentmanager) with new `share_items()` and `unshare_items()` functions
* Added `history()` function to [`ArcGISOnlineManager`](https://developers.arcgis.com/python/api-reference/arcgis.gis.admin/#agoladminmanager) and [`PortalAdminManager`](https://developers.arcgis.com/python/api-reference/arcgis.gis.admin/#portaladminmanager) objects to review logins to a portal
* Added support to use [`Spatially Enabled DataFrames`](https://developers.arcgis.com/python/api-reference/arcgis.features.toc/#arcgis-features-geoaccessor) as `study_area` parameters on the [`geoenrichment.enrich()`](https://developers.arcgis.com/python/api-reference/arcgis.geoenrichment/#enrich) function
* Added `take_screenshot()` function to the [`MapWidget`](https://developers.arcgis.com/python/api-reference/arcgis.widgets/#mapview) for use on [`map`](https://developers.arcgis.com/python/api-reference/arcgis.geoenrichment/#enrich) objects
* Fixed the [`Item.usage()`](https://developers.arcgis.com/python/api-reference/arcgis.gis.toc/#arcgis.gis.Item.usage) to correctly process `6M` and `1Y` input date ranges
* Fixed [`StoryMap.save()`](https://developers.arcgis.com/python/api-reference/arcgis.apps.storymap/#arcgis.apps.storymap.JournalStoryMap.save) method for when `gis` is ArcGIS Enterprise
* Improved various [`features.analysis`](https://developers.arcgis.com/python/api-reference/arcgis.features.analysis/) tools to accept lists of [`FeatureLayers`](https://developers.arcgis.com/python/api-reference/arcgis.features.toc/#featurelayer) as inputs
* Fixed [`add_layer()`](https://developers.arcgis.com/python/api-reference/arcgis.widgets/#arcgis.widgets.MapView.add_layer) to properly render layers with definition expressions
* Fixed issue with [`geometry.filters`](https://developers.arcgis.com/python/api-reference/arcgis.geometry.filters/) incorrectly processing various geometry types
* Fixed [`ServerManager.list()`]() function to improve reading of returned [`servers`](https://developers.arcgis.com/python/api-reference/arcgis.gis.server/#server)
* Fixed Spatially Enabled DataFrame [`from_layer()`](https://developers.arcgis.com/python/api-reference/arcgis.features.toc/#arcgis.features.GeoAccessor.from_layer) function to better read date input
* Fixed issue with [`clear_graphics()`](https://developers.arcgis.com/python/api-reference/arcgis.widgets/#arcgis.widgets.MapView.clear_graphics) function not removing layers when run consecutively
* Fixed issue with [`Item`](https://developers.arcgis.com/python/api-reference/arcgis.gis.toc/#item) thumbnails not rendring properly under specific circumstances in Jupyter notebooks
* Improved handling of `date` fields in the [`query()`](https://developers.arcgis.com/python/api-reference/arcgis.features.toc/#arcgis.features.FeatureLayer.query) method
